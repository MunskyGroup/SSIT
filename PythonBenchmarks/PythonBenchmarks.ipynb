```python
modelGoutsiasRebop = rebop.Gillespie()

# Parameters
paramsGoutsiasRebop = {
    "k1": 0.043,
    "k2": 0.0007,
    "k3": 0.0715,
    "k4": 0.0039,
    "k5": 0.012e9/(6.0221415e23*1e-15),
    "k6": 0.4791,
    "k7": 0.00012e9/(6.0221415e23*1e-15),
    "k8": 0.8765e-11,
    "k9": 0.05e9/(6.0221415e23*1e-15)/2,
    "k10": 0.5
}

# Species:
# D, DNA, M, RNA, DNAD, DNA2D

# 1. RNA -> RNA + M
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k1"], ["RNA"], ["RNA", "M"])

# 2. M -> empty
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k2"], ["M"], [])

# 3. DNAD -> DNAD + RNA
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k3"], ["DNAD"], ["DNAD", "RNA"])

# 4. RNA -> empty
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k4"], ["RNA"], [])

# 5. DNA + D -> DNAD
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k5"], ["DNA", "D"], ["DNAD"])

# 6. DNAD -> DNA + D
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k6"], ["DNAD"], ["DNA", "D"])

# 7. DNAD + D -> DNA2D
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k7"], ["DNAD", "D"], ["DNA2D"])

# 8. DNA2D -> DNAD + D
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k8"], ["DNA2D"], ["DNAD", "D"])

# 9. 2M -> D
# propensity = (k9/2)*M*(M-1)
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k9"], ["M", "M"], ["D"])

# 10. D -> 2M
modelGoutsiasRebop.add_reaction(paramsGoutsiasRebop["k10"], ["D"], ["M", "M"])

# Initial conditions
x0GoutsiasRebop = {
    "D": 6,
    "DNA": 2,
    "M": 2,
    "RNA": 0,
    "DNAD": 0,
    "DNA2D": 0
}

# Run Once
ds = modelGoutsiasRebop.run(
    x0GoutsiasRebop,
    params=paramsGoutsiasRebop,
    tmax=100,
    nb_steps=101
)
```

```python
# Michaelis-Menten model in ReBop
modelMMRebop = rebop.Gillespie()

# Parameters
paramsMMRebop = {"k1": 1.0, "k2": 1.0, "k3": 0.1}

# Species: S, E, ES, P
# Reactions from stoichiometry/propensity definitions:
# 1) S + E -> ES   (k1*S*E)
# 2) ES -> S + E   (k2*ES)
# 3) ES -> E + P   (k3*ES)
modelMMRebop.add_reaction(paramsMMRebop["k1"], ["S", "E"], ["ES"])
modelMMRebop.add_reaction(paramsMMRebop["k2"], ["ES"], ["S", "E"])
modelMMRebop.add_reaction(paramsMMRebop["k3"], ["ES"], ["E", "P"])

# Example initial conditions (edit as needed)
x0MMRebop = {"S": 100, "E": 30, "ES": 0, "P": 0}

# Quick test run
dsMM = modelMMRebop.run(x0MMRebop, params=paramsMMRebop, tmax=50, nb_steps=51)
```

```python
# Birth-death (BD) model in ReBop
modelBDRebop = rebop.Gillespie()

# Parameters
paramsBDRebop = {"bk": 1000.0, "dk": 1.0}

# Species: X
# Reactions from stoichiometry/propensity definitions:
# 1) Birth:   empty -> X   with propensity bk
# 2) Death:   X -> empty   with propensity dk*X
modelBDRebop.add_reaction(paramsBDRebop["bk"], [], ["X"])
modelBDRebop.add_reaction(paramsBDRebop["dk"], ["X"], [])

# Initial condition
x0BDRebop = {"X": 0}

# Quick test run
dsBD = modelBDRebop.run(x0BDRebop, params=paramsBDRebop, tmax=10, nb_steps=11)
```

```python
# MAPK model in ReBop
modelMAPKRebop = rebop.Gillespie()

# Parameters
paramsMAPKRebop = {
    "s1": 0.00024,
    "d1": 0.0001,
    "s2": 0.001,
    "d2": 0.15,
    "s3": 0.005,
    "k1": 0.375,
    "k_1": 1.0,
    "k2": 0.06,
    "k3": 0.375,
    "k_3": 1.0,
    "k4": 4.5,
    "k5": 0.375,
    "k_5": 1.0,
    "k6": 0.06,
    "k7": 0.375,
    "k_7": 1.0,
    "k8": 4.5,
    "h1": 0.015,
    "h_1": 1.0,
    "h2": 0.032,
    "h3": 0.31,
    "h_3": 0.01,
    "h4": 0.01,
    "h_4": 1.0,
    "h5": 0.5,
    "h6": 0.086,
    "h_6": 0.0011,
    "h7": 0.01,
    "h_7": 1.0,
    "h8": 0.47,
    "h9": 0.14,
    "h_9": 0.0018,
}

# Species initial conditions
x0MAPKRebop = {
    "MKP3": 1,
    "Kpp_MKP3": 0,
    "KpT_MKP3_Y": 0,
    "KpT_MKP3_T": 0,
    "K_MKP3_T": 0,
    "KpY_MKP3": 0,
    "K_MKP3_Y": 0,
    "KpT_MEK": 0,
    "KpT": 0,
    "KpY_MEK": 0,
    "K_MEK_T": 0,
    "KpY": 0,
    "K_MEK_Y": 0,
    "K": 3,
    "MEK": 0,
    "Kpp": 0,
}

# R1: synthesis/degradation of MEK
modelMAPKRebop.add_reaction(paramsMAPKRebop["s2"], [], ["MEK"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["d2"], ["MEK"], [])

# R2: Kpp-driven synthesis of MEK
modelMAPKRebop.add_reaction(paramsMAPKRebop["s3"], ["Kpp"], ["Kpp", "MEK"])

# R3: synthesis/degradation of K
modelMAPKRebop.add_reaction(paramsMAPKRebop["s1"], [], ["K"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["d1"], ["K"], [])

# R4: degradation of phosphorylated K forms
modelMAPKRebop.add_reaction(paramsMAPKRebop["d1"], ["KpY"], [])
modelMAPKRebop.add_reaction(paramsMAPKRebop["d1"], ["KpT"], [])
modelMAPKRebop.add_reaction(paramsMAPKRebop["d1"], ["Kpp"], [])

# R5: K + MEK -> K_MEK_Y
modelMAPKRebop.add_reaction(paramsMAPKRebop["k1"], ["K", "MEK"], ["K_MEK_Y"])

# R6: K_MEK_Y -> KpY + MEK
modelMAPKRebop.add_reaction(paramsMAPKRebop["k2"], ["K_MEK_Y"], ["KpY", "MEK"])

# R7: KpY + MEK <=> KpY_MEK
modelMAPKRebop.add_reaction(paramsMAPKRebop["k3"], ["KpY", "MEK"], ["KpY_MEK"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["k_3"], ["KpY_MEK"], ["KpY", "MEK"])

# R8: KpY_MEK -> Kpp + MEK
modelMAPKRebop.add_reaction(paramsMAPKRebop["k4"], ["KpY_MEK"], ["Kpp", "MEK"])

# R9: K + MEK <=> K_MEK_T
modelMAPKRebop.add_reaction(paramsMAPKRebop["k5"], ["K", "MEK"], ["K_MEK_T"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["k_5"], ["K_MEK_T"], ["K", "MEK"])

# R10: K_MEK_T -> KpT + MEK
modelMAPKRebop.add_reaction(paramsMAPKRebop["k6"], ["K_MEK_T"], ["KpT", "MEK"])

# R11: KpT + MEK <=> KpT_MEK
modelMAPKRebop.add_reaction(paramsMAPKRebop["k7"], ["KpT", "MEK"], ["KpT_MEK"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["k_7"], ["KpT_MEK"], ["KpT", "MEK"])

# R12: KpT_MEK -> Kpp + MEK
modelMAPKRebop.add_reaction(paramsMAPKRebop["k8"], ["KpT_MEK"], ["Kpp", "MEK"])

# R13: Kpp + MKP3 <=> Kpp_MKP3
modelMAPKRebop.add_reaction(paramsMAPKRebop["h1"], ["Kpp", "MKP3"], ["Kpp_MKP3"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["h_1"], ["Kpp_MKP3"], ["Kpp", "MKP3"])

# R14: Kpp_MKP3 -> KpT_MKP3_Y
modelMAPKRebop.add_reaction(paramsMAPKRebop["h2"], ["Kpp_MKP3"], ["KpT_MKP3_Y"])

# R15: KpT_MKP3_Y <=> KpT + MKP3
modelMAPKRebop.add_reaction(paramsMAPKRebop["h3"], ["KpT_MKP3_Y"], ["KpT", "MKP3"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["h_3"], ["KpT", "MKP3"], ["KpT_MKP3_Y"])

# R16: KpT + MKP3 <=> KpT_MKP3_T
modelMAPKRebop.add_reaction(paramsMAPKRebop["h4"], ["KpT", "MKP3"], ["KpT_MKP3_T"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["h_4"], ["KpT_MKP3_T"], ["KpT", "MKP3"])

# R17: KpT_MKP3_T -> K_MKP3_T
modelMAPKRebop.add_reaction(paramsMAPKRebop["h5"], ["KpT_MKP3_T"], ["K_MKP3_T"])

# R18: K_MKP3_T <=> K + MKP3
modelMAPKRebop.add_reaction(paramsMAPKRebop["h6"], ["K_MKP3_T"], ["K", "MKP3"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["h_6"], ["K", "MKP3"], ["K_MKP3_T"])

# R19: KpY + MKP3 <=> KpY_MKP3
modelMAPKRebop.add_reaction(paramsMAPKRebop["h7"], ["KpY", "MKP3"], ["KpY_MKP3"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["h_7"], ["KpY_MKP3"], ["KpY", "MKP3"])

# R20: KpY_MKP3 -> K_MKP3_Y
modelMAPKRebop.add_reaction(paramsMAPKRebop["h8"], ["KpY_MKP3"], ["K_MKP3_Y"])

# R21: K_MKP3_Y <=> K + MKP3
modelMAPKRebop.add_reaction(paramsMAPKRebop["h9"], ["K_MKP3_Y"], ["K", "MKP3"])
modelMAPKRebop.add_reaction(paramsMAPKRebop["h_9"], ["K", "MKP3"], ["K_MKP3_Y"])

# Quick test run
dsMAPK = modelMAPKRebop.run(x0MAPKRebop, params=paramsMAPKRebop, tmax=10, nb_steps=101)
```

```python
# ReBop model for 6-species toggle model
modelToggleRebop = rebop.Gillespie()

# Parameters
paramsToggleRebop = {
    "k1": 40.0,
    "k2": 20.0,
    "k3": 1.0,
    "k4": 1.0,
    "k5": 1e-5/2,
    "k6": 3.5e-5/2,
    "k7": 1.0,
    "k8": 1.0,
}

# Species order: GeneA, GeneB, A, B, bGeneB, bGeneA
x0ToggleRebop = {
    "GeneA": 1,
    "GeneB": 1,
    "A": 0,
    "B": 0,
    "bGeneB": 0,
    "bGeneA": 0,
}

# R1: A production from GeneA: k1*GeneA
modelToggleRebop.add_reaction(paramsToggleRebop["k1"], ["GeneA"], ["GeneA", "A"])

# R2: B production from GeneB: k2*GeneB
modelToggleRebop.add_reaction(paramsToggleRebop["k2"], ["GeneB"], ["GeneB", "B"])

# R3-R4: degradation
modelToggleRebop.add_reaction(paramsToggleRebop["k3"], ["A"], [])
modelToggleRebop.add_reaction(paramsToggleRebop["k4"], ["B"], [])

# R5 and R7: 2A + GeneB <=> bGeneB
modelToggleRebop.add_reaction(paramsToggleRebop["k5"], ["A", "A", "GeneB"], ["bGeneB"])
modelToggleRebop.add_reaction(paramsToggleRebop["k7"], ["bGeneB"], ["A", "A", "GeneB"])

# R6 and R8: 2B + GeneA <=> bGeneA
modelToggleRebop.add_reaction(paramsToggleRebop["k6"], ["B", "B", "GeneA"], ["bGeneA"])
modelToggleRebop.add_reaction(paramsToggleRebop["k8"], ["bGeneA"], ["B", "B", "GeneA"])

# Quick test run (timeSets = 30)
dsToggle = modelToggleRebop.run(x0ToggleRebop, params=paramsToggleRebop, tmax=30, nb_steps=11)
```

```python
import gillespy2
import numpy as np

class GoutsiasModelGillesPy2(gillespy2.Model):

    def __init__(self):
        super().__init__(name="DNA_expression")

        # Parameters
        k1  = gillespy2.Parameter(name="k1",  expression=0.043)
        k2  = gillespy2.Parameter(name="k2",  expression=0.0007)
        k3  = gillespy2.Parameter(name="k3",  expression=0.0715)
        k4  = gillespy2.Parameter(name="k4",  expression=0.0039)
        k5  = gillespy2.Parameter(name="k5",  expression=0.012e9/(6.0221415e23*1e-15))
        k6  = gillespy2.Parameter(name="k6",  expression=0.4791)
        k7  = gillespy2.Parameter(name="k7",  expression=0.00012e9/(6.0221415e23*1e-15))
        k8  = gillespy2.Parameter(name="k8",  expression=0.8765e-11)
        k9  = gillespy2.Parameter(name="k9",  expression=0.05e9/(6.0221415e23*1e-15)/2)
        k10 = gillespy2.Parameter(name="k10", expression=0.5)

        self.add_parameter([
            k1,k2,k3,k4,k5,k6,k7,k8,k9,k10
        ])

        # Species
        D     = gillespy2.Species(name="D",     initial_value=6)
        DNA   = gillespy2.Species(name="DNA",   initial_value=2)
        M     = gillespy2.Species(name="M",     initial_value=2)
        RNA   = gillespy2.Species(name="RNA",   initial_value=0)
        DNAD  = gillespy2.Species(name="DNAD",  initial_value=0)
        DNA2D = gillespy2.Species(name="DNA2D", initial_value=0)

        self.add_species([
            D, DNA, M, RNA, DNAD, DNA2D
        ])

        # Reactions
        reactions = [

            # 1. RNA -> RNA + M
            gillespy2.Reaction(
                name="translation",
                rate=k1,
                reactants={RNA:1},
                products={RNA:1, M:1}
            ),

            # 2. M -> ∅
            gillespy2.Reaction(
                name="M_degradation",
                rate=k2,
                reactants={M:1},
                products={}
            ),

            # 3. DNAD -> DNAD + RNA
            gillespy2.Reaction(
                name="RNA_production",
                rate=k3,
                reactants={DNAD:1},
                products={DNAD:1, RNA:1}
            ),

            # 4. RNA -> ∅
            gillespy2.Reaction(
                name="RNA_degradation",
                rate=k4,
                reactants={RNA:1},
                products={}
            ),

            # 5. DNA + D -> DNAD
            gillespy2.Reaction(
                name="DNA_binding",
                rate=k5,
                reactants={DNA:1, D:1},
                products={DNAD:1}
            ),

            # 6. DNAD -> DNA + D
            gillespy2.Reaction(
                name="DNA_unbinding",
                rate=k6,
                reactants={DNAD:1},
                products={DNA:1, D:1}
            ),

            # 7. DNAD + D -> DNA2D
            gillespy2.Reaction(
                name="second_D_binding",
                rate=k7,
                reactants={DNAD:1, D:1},
                products={DNA2D:1}
            ),

            # 8. DNA2D -> DNAD + D
            gillespy2.Reaction(
                name="second_D_unbinding",
                rate=k8,
                reactants={DNA2D:1},
                products={DNAD:1, D:1}
            ),

            # 9. 2M -> D
            # Need custom propensity to reproduce (k9/2)*M*(M-1)
            gillespy2.Reaction(
                name="M_dimerization",
                rate=k9,
                # propensity_function="(k9/2)*M*(M-1)",
                reactants={M:2},
                products={D:1}
            ),

            # 10. D -> 2M
            gillespy2.Reaction(
                name="D_dissociation",
                rate=k10,
                reactants={D:1},
                products={M:2}
            )
        ]

        self.add_reaction(reactions)

        # Time span
        self.timespan(
            gillespy2.TimeSpan.linspace(
                t = 100,
                num_points=11
            )
        )


# Create and run modelGoutsiasGillesPy2
modelGoutsiasGillesPy2 = GoutsiasModelGillesPy2()
```

```python
# MM model in GillesPy2
class MMModelGillesPy2(gillespy2.Model):

    def __init__(self):
        super().__init__(name="MM_model")

        # Parameters
        k1 = gillespy2.Parameter(name="k1", expression=1.0)
        k2 = gillespy2.Parameter(name="k2", expression=1.0)
        k3 = gillespy2.Parameter(name="k3", expression=0.1)

        self.add_parameter([k1, k2, k3])

        # Species
        S  = gillespy2.Species(name="S",  initial_value=100)
        E  = gillespy2.Species(name="E",  initial_value=30)
        ES = gillespy2.Species(name="ES", initial_value=0)
        P  = gillespy2.Species(name="P",  initial_value=0)

        self.add_species([S, E, ES, P])

        # Reactions
        reactions = [
            # 1) S + E -> ES   (k1*S*E)
            gillespy2.Reaction(
                name="binding",
                rate=k1,
                reactants={S:1, E:1},
                products={ES:1}
            ),

            # 2) ES -> S + E   (k2*ES)
            gillespy2.Reaction(
                name="unbinding",
                rate=k2,
                reactants={ES:1},
                products={S:1, E:1}
            ),

            # 3) ES -> E + P   (k3*ES)
            gillespy2.Reaction(
                name="catalysis",
                rate=k3,
                reactants={ES:1},
                products={E:1, P:1}
            )
        ]

        self.add_reaction(reactions)

        # Time span
        self.timespan(
            gillespy2.TimeSpan.linspace(
                t = 70,
                num_points=11
            )
        )
```

```python
# Birth-death model in GillesPy2
class BDModelGillesPy2(gillespy2.Model):

    def __init__(self):
        super().__init__(name="BD_model")

        # Parameters
        bk = gillespy2.Parameter(name="bk", expression=1000.0)
        dk = gillespy2.Parameter(name="dk", expression=1.0)

        self.add_parameter([bk, dk])

        # Species
        X = gillespy2.Species(name="X", initial_value=0)

        self.add_species([X])

        # Reactions
        reactions = [
            # 1) Birth:   ∅ -> X   with propensity bk
            gillespy2.Reaction(
                name="birth",
                rate=bk,
                reactants={},
                products={X:1}
            ),

            # 2) Death:   X -> ∅   with propensity dk*X
            gillespy2.Reaction(
                name="death",
                rate=dk,
                reactants={X:1},
                products={}
            )
        ]

        self.add_reaction(reactions)

        # Time span
        self.timespan(
            gillespy2.TimeSpan.linspace(
                t = 10,
                num_points=11
            )
        )
```

```python
# 6-species toggle model in GillesPy2
class ToggleModelGillesPy2(gillespy2.Model):

    def __init__(self):
        super().__init__(name="toggle_6_species")

        # Parameters
        p = {
            "k1": gillespy2.Parameter(name="k1", expression=40.0),
            "k2": gillespy2.Parameter(name="k2", expression=20.0),
            "k3": gillespy2.Parameter(name="k3", expression=1.0),
            "k4": gillespy2.Parameter(name="k4", expression=1.0),
            "k5": gillespy2.Parameter(name="k5", expression=1e-5),
            "k6": gillespy2.Parameter(name="k6", expression=3.5e-5),
            "k7": gillespy2.Parameter(name="k7", expression=1.0),
            "k8": gillespy2.Parameter(name="k8", expression=1.0),
        }
        self.add_parameter(list(p.values()))

        # Species initial conditions
        s = {
            "GeneA": gillespy2.Species(name="GeneA", initial_value=1),
            "GeneB": gillespy2.Species(name="GeneB", initial_value=1),
            "A": gillespy2.Species(name="A", initial_value=0),
            "B": gillespy2.Species(name="B", initial_value=0),
            "bGeneB": gillespy2.Species(name="bGeneB", initial_value=0),
            "bGeneA": gillespy2.Species(name="bGeneA", initial_value=0),
        }
        self.add_species(list(s.values()))

        # Reactions
        reactions = [
            # k1*GeneA
            gillespy2.Reaction(name="r1", rate=p["k1"], reactants={s["GeneA"]: 1}, products={s["GeneA"]: 1, s["A"]: 1}),
            # k2*GeneB
            gillespy2.Reaction(name="r2", rate=p["k2"], reactants={s["GeneB"]: 1}, products={s["GeneB"]: 1, s["B"]: 1}),
            # k3*A
            gillespy2.Reaction(name="r3", rate=p["k3"], reactants={s["A"]: 1}, products={}),
            # k4*B
            gillespy2.Reaction(name="r4", rate=p["k4"], reactants={s["B"]: 1}, products={}),
            # (k5/2)*A*(A-1)*GeneB
            gillespy2.Reaction(
                name="r5",
                reactants={s["A"]: 2, s["GeneB"]: 1},
                products={s["bGeneB"]: 1},
                propensity_function="(k5/2.0)*A*(A-1)*GeneB"
            ),
            # (k6/2)*B*(B-1)*GeneA
            gillespy2.Reaction(
                name="r6",
                reactants={s["B"]: 2, s["GeneA"]: 1},
                products={s["bGeneA"]: 1},
                propensity_function="(k6/2.0)*B*(B-1)*GeneA"
            ),
            # k7*bGeneB
            gillespy2.Reaction(name="r7", rate=p["k7"], reactants={s["bGeneB"]: 1}, products={s["A"]: 2, s["GeneB"]: 1}),
            # k8*bGeneA
            gillespy2.Reaction(name="r8", rate=p["k8"], reactants={s["bGeneA"]: 1}, products={s["B"]: 2, s["GeneA"]: 1}),
        ]
        self.add_reaction(reactions)

        # timeSets = 30
        self.timespan(gillespy2.TimeSpan.linspace(t=30, num_points=31))


modelToggleGillesPy2 = ToggleModelGillesPy2()
```

```python
# Phage model in GillesPy2
class PhageModelGillesPy2(gillespy2.Model):

    def __init__(self):
        super().__init__(name="phage_model")

        # Parameters
        pvals = {
            "k1": 0.0069, "k2": 0.0069, "k3": 0.069, "k4": 0.0929, "k5": 0.0026,
            "k6": 0.0025, "k7": 0.021, "k8": 0.021, "k9": 0.021, "k10": 0.021,
            "k11": 0.021, "k12": 0.021, "k13": 0.00898, "k14": 0.00011, "k15": 0.01242,
            "k16": 0.00011, "k17": 0.00898, "k18": 0.2297, "k19": 0.0029, "k20": 0.0021,
            "k21": 0.0029, "k22": 0.2297, "k23": 0.2297, "k24": 0.2297, "k25": 0.0029,
            "k26": 0.0021, "k27": 1.13, "k28": 0.0106, "k29": 0.0106, "k30": 0.0106,
            "k31": 1.13, "k32": 0.0202, "k33": 0.0202, "k34": 0.0040, "k35": 0.0040,
            "k36": 0.0040, "k37": 0.1413, "k38": 0.1413, "k39": 0.1413, "k40": 0.1413,
            "k41": 0.0279, "k42": 0.053, "k43": 0.0328, "k44": 0.053, "k45": 0.0279,
            "k46": 0.0022, "k47": 0.0022, "k48": 0.0008, "k49": 0.0008, "k50": 0.003,
        }
        params = {k: gillespy2.Parameter(name=k, expression=v) for k, v in pvals.items()}
        self.add_parameter(list(params.values()))

        # Species initial conditions
        s0 = {
            "OR1": 1, "OR2": 1, "OR3": 1,
            "COR1": 0, "COR2": 0, "COR3": 0,
            "ROR1": 0, "ROR2": 0, "ROR3": 0,
            "CI2": 0, "Cro2": 0,
        }
        s = {name: gillespy2.Species(name=name, initial_value=val) for name, val in s0.items()}
        self.add_species(list(s.values()))

        # (propensity, stoichiometry dict)
        reaction_data = [
            ("k1*OR3*OR2", {"CI2": 1}),
            ("k2*OR3*COR2", {"CI2": 1}),
            ("k3*OR3*ROR2", {"CI2": 1}),
            ("k4*OR1*OR2", {"Cro2": 1}),
            ("k5*CI2", {"CI2": -1}),
            ("k6*Cro2", {"Cro2": -1}),
            ("k7*CI2*OR1", {"CI2": -1, "OR1": -1, "ROR1": 1}),
            ("k8*CI2*OR2", {"CI2": -1, "OR2": -1, "ROR2": 1}),
            ("k9*CI2*OR3", {"CI2": -1, "OR3": -1, "ROR3": 1}),
            ("k10*Cro2*OR1", {"Cro2": -1, "OR1": -1, "COR1": 1}),
            ("k11*Cro2*OR2", {"Cro2": -1, "OR2": -1, "COR2": 1}),
            ("k12*Cro2*OR3", {"Cro2": -1, "OR3": -1, "COR3": 1}),
            ("k13*ROR1*OR2", {"ROR1": -1, "CI2": 1, "OR1": 1}),
            ("k14*ROR1*ROR2*OR3", {"ROR1": -1, "CI2": 1, "OR1": 1}),
            ("k15*ROR1*ROR2*ROR3", {"ROR1": -1, "CI2": 1, "OR1": 1}),
            ("k16*ROR1*ROR2*COR3", {"ROR1": -1, "CI2": 1, "OR1": 1}),
            ("k17*ROR1*COR2", {"ROR1": -1, "CI2": 1, "OR1": 1}),
            ("k18*ROR2*OR1*OR3", {"ROR2": -1, "CI2": 1, "OR2": 1}),
            ("k19*ROR2*ROR1*OR3", {"ROR2": -1, "CI2": 1, "OR2": 1}),
            ("k20*ROR2*OR1*ROR3", {"ROR2": -1, "CI2": 1, "OR2": 1}),
            ("k21*ROR2*ROR1*ROR3", {"ROR2": -1, "CI2": 1, "OR2": 1}),
            ("k22*ROR2*COR1*OR3", {"ROR2": -1, "CI2": 1, "OR2": 1}),
            ("k23*ROR2*OR1*COR3", {"ROR2": -1, "CI2": 1, "OR2": 1}),
            ("k24*ROR2*COR1*COR3", {"ROR2": -1, "CI2": 1, "OR2": 1}),
            ("k25*ROR2*ROR1*COR3", {"ROR2": -1, "CI2": 1, "OR2": 1}),
            ("k26*ROR2*COR1*ROR3", {"ROR2": -1, "CI2": 1, "OR2": 1}),
            ("k27*ROR3*OR2", {"ROR3": -1, "CI2": 1, "OR3": 1}),
            ("k28*ROR3*ROR2*OR1", {"ROR3": -1, "CI2": 1, "OR3": 1}),
            ("k29*ROR3*ROR2*ROR1", {"ROR3": -1, "CI2": 1, "OR3": 1}),
            ("k30*ROR3*ROR2*COR1", {"ROR3": -1, "CI2": 1, "OR3": 1}),
            ("k31*ROR3*COR2", {"ROR3": -1, "CI2": 1, "OR3": 1}),
            ("k32*COR1*OR2", {"COR1": -1, "Cro2": 1, "OR1": 1}),
            ("k33*COR1*ROR2", {"COR1": -1, "Cro2": 1, "OR1": 1}),
            ("k34*COR1*COR2*OR3", {"COR1": -1, "Cro2": 1, "OR1": 1}),
            ("k35*COR1*COR2*ROR3", {"COR1": -1, "Cro2": 1, "OR1": 1}),
            ("k36*COR1*COR2*COR3", {"COR1": -1, "Cro2": 1, "OR1": 1}),
            ("k37*COR2*OR1*OR3", {"COR2": -1, "Cro2": 1, "OR2": 1}),
            ("k38*COR2*ROR1*OR3", {"COR2": -1, "Cro2": 1, "OR2": 1}),
            ("k39*COR2*OR1*ROR3", {"COR2": -1, "Cro2": 1, "OR2": 1}),
            ("k40*COR2*ROR1*ROR3", {"COR2": -1, "Cro2": 1, "OR2": 1}),
            ("k41*COR2*COR1*OR3", {"COR2": -1, "Cro2": 1, "OR2": 1}),
            ("k42*COR2*OR1*COR3", {"COR2": -1, "Cro2": 1, "OR2": 1}),
            ("k43*COR2*COR1*COR3", {"COR2": -1, "Cro2": 1, "OR2": 1}),
            ("k44*COR2*ROR1*COR3", {"COR2": -1, "Cro2": 1, "OR2": 1}),
            ("k45*COR2*COR1*ROR3", {"COR2": -1, "Cro2": 1, "OR2": 1}),
            ("k46*COR3*OR2", {"COR3": -1, "Cro2": 1, "OR3": 1}),
            ("k47*COR3*ROR2", {"COR3": -1, "Cro2": 1, "OR3": 1}),
            ("k48*COR3*COR2*OR1", {"COR3": -1, "Cro2": 1, "OR3": 1}),
            ("k49*COR3*COR2*ROR1", {"COR3": -1, "Cro2": 1, "OR3": 1}),
            ("k50*COR3*COR2*COR1", {"COR3": -1, "Cro2": 1, "OR3": 1}),
        ]

        reactions = []
        for i, (prop, stoich) in enumerate(reaction_data, start=1):
            reactants = {}
            products = {}
            for sp, coeff in stoich.items():
                if coeff < 0:
                    reactants[s[sp]] = -coeff
                elif coeff > 0:
                    products[s[sp]] = coeff
            reactions.append(
                gillespy2.Reaction(
                    name=f"r{i}",
                    reactants=reactants,
                    products=products,
                    propensity_function=prop,
                )
            )

        self.add_reaction(reactions)
        self.timespan(gillespy2.TimeSpan.linspace(t=30, num_points=31))


modelPhageGillesPy2 = PhageModelGillesPy2()
```
