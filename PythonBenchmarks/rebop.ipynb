```python
import rebop
import time
```

```python

sir = rebop.Gillespie()
sir.add_reaction(1e-4, ['S', 'I'], ['I', 'I'])
sir.add_reaction(0.01, ['I'], ['R'])
print(sir)
```

```python
ds = sir.run({'S': 999, 'I': 1}, tmax=250, nb_steps=25)
```

```python
startTime = time.time()

for i in range(1000):
    ds = sir.run({'S': 999, 'I': 1}, tmax=250, nb_steps=25)

print("--- %s seconds ---" % (time.time() - startTime))
```
